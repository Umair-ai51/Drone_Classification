```python
X = df_combined[['Time', ' OrgTrans_X', ' ModTrans_X', ' OrgTrans_Y', ' ModTrans_Y',
       ' OrgTrans_Z', ' ModTrans_Z', ' OrgRot_w', ' ModRot_w', ' OrgRot_x',
       'ModRot_x', ' OrgRot_y', ' ModRot_y', ' OrgRot_z ', 'ModRot_z',
       'Radar_Velocity', 'Radar_Range', 'Acoustic_Signal', 'Tracking_Error',
       ' Optical_Error', ' VBAKF_Error', ' vbakf_x', ' vbakf_y', ' optical_x',
       ' optical_y']]

y = df_combined['Type']
```

```python
core_features = ['Time', ' OrgTrans_X', ' ModTrans_X', ' OrgTrans_Y', ' ModTrans_Y',
       ' OrgTrans_Z', ' ModTrans_Z', ' OrgRot_w', ' ModRot_w', ' OrgRot_x',
       'ModRot_x', ' OrgRot_y', ' ModRot_y', ' OrgRot_z ', 'ModRot_z',
       'Radar_Velocity', 'Radar_Range', 'Acoustic_Signal', 'Tracking_Error',
       ' Optical_Error', ' VBAKF_Error', ' vbakf_x', ' vbakf_y', ' optical_x',
       ' optical_y']

```

```python
def prepare_sequences(X, y, sequence_length=10):
    """
    Prepare sequences with overlap for LSTM and adjust y accordingly.
    """
    features = []
    targets = []  # Store corresponding targets
    step = 2  # Use overlapping sequences
    for i in range(0, len(X) - sequence_length, step):
        features.append(X[i:(i + sequence_length)])
        targets.append(y[i + sequence_length -1])  # Get the target for the last element of the sequence

    return np.array(features), np.array(targets)
```

```python
X_core = X[core_features]

# Scale the features
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X_core)

# Prepare sequences, also prepare y_sequences
sequence_length = 8
X_sequences, y_sequences = prepare_sequences(X_scaled, y.values, sequence_length)  # pass both X and y to function

# Split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_sequences, y_sequences,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=y_sequences
)
```

```python
def create_lstm_model(input_shape):
    model = Sequential([
        # First Bidirectional LSTM layer
        Bidirectional(LSTM(64, return_sequences=True),
                     input_shape=input_shape),
        BatchNormalization(),
        Dropout(0.3),

        # Second Bidirectional LSTM layer
        Bidirectional(LSTM(32)),
        BatchNormalization(),
        Dropout(0.3),

        # Dense layers with batch normalization
        Dense(32, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(16, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
input_shape = (X_train.shape[1], X_train.shape[2])
model = create_lstm_model(input_shape)

# Callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    min_delta=0.001
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=3,
    min_lr=0.00001
)

# Train with class weights
class_weights = dict(zip(
    np.unique(y_train),
    1 / np.bincount(y_train) * len(y_train) / 2
))

history = model.fit(
    X_train, y_train,
    epochs=40,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping, reduce_lr],
    class_weight=class_weights,
    verbose=1
)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"\nTest accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype(int)

# Print detailed classification report
print("Classification Report:")
print(classification_report(y_test, y_pred_binary))
```
